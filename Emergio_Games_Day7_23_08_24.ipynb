32. Given a time series DataFrame 'df', resample the data to a monthly frequency and fill any missing months with the forward-fill method.

```python
def resample_to_monthly_and_ffill(df):
    df.index = pd.to_datetime(df.index)

    df_resampled = df.resample('M').ffill()

    return df_resampled
```

35. Write a function to identify and remove outliers from a DataFrame 'df' using the Z-score method.

```python
import pandas as pd
import numpy as np
from scipy import stats

def remove_outliers_zscore(df, threshold=3):
    df_numerical = df.select_dtypes(include=[np.number])

    z_scores = np.abs(stats.zscore(df_numerical))

    mask = (z_scores < threshold).all(axis=1)

    df_cleaned = df[mask]

    return df_cleaned

data = {
    'A': [10, 20, 30, 1000],
    'B': [5, 10, 15, -500],
    'C': [7, 8, 9, 100]
}
df = pd.DataFrame(data)

df_no_outliers = remove_outliers_zscore(df)
print(df_no_outliers)
```

39. Write a function to perform one-hot encoding on the categorical columns of a DataFrame 'df'.

```python
def one_hot_encode(df):
    categorical_cols = df.select_dtypes(include=['object', 'category']).columns

    df_encoded = pd.get_dummies(df, columns=categorical_cols)

    return df_encoded

data = {
    'department': ['HR', 'Finance', 'HR', 'Finance', 'IT'],
    'gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'salary': [50000, 60000, 55000, 62000, 70000]
}
df = pd.DataFrame(data)

df_encoded = one_hot_encode(df)
print(df_encoded)
```
