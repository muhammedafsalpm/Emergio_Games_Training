10. Use Matplotlib to create a subplot with four different plots: a line plot, a bar chart, a histogram, and a scatter plot.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.random.randint(0, 10, 10)
y3 = np.random.normal(5, 2, 100)
y4 = np.random.rand(100)

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 8))

# Plot 1: Line plot
ax1.plot(x, y1, color='blue')
ax1.set_title('Line Plot')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')

# Plot 2: Bar chart
ax2.bar(range(len(y2)), y2, color='orange')
ax2.set_title('Bar Chart')
ax2.set_xlabel('X-axis')
ax2.set_ylabel('Y-axis')

# Plot 3: Histogram
ax3.hist(y3, bins=10, color='green')
ax3.set_title('Histogram')
ax3.set_xlabel('X-axis')
ax3.set_ylabel('Frequency')

# Plot 4: Scatter plot
ax4.scatter(y4, y3, color='purple')
ax4.set_title('Scatter Plot')
ax4.set_xlabel('X-axis')
ax4.set_ylabel('Y-axis')

# Adjust spacing between subplots
plt.tight_layout()

plt.show()
```

19. Create a Python program to plot a radar chart (spider plot) using Matplotlib to compare the performance metrics of multiple entities.

```python
import matplotlib.pyplot as plt
import numpy as np

labels = ['Category 1', 'Category 2', 'Category 3', 'Category 4', 'Category 5']
data = [[3, 4, 5, 2, 4],
        [2, 3, 4, 5, 3],
        [4, 5, 3, 2, 4]]

data = np.array(data)

num_categories = len(labels)
angles = np.linspace(0, 2 * np.pi, num_categories, endpoint=False)

angles = np.append(angles, angles[0])

data = np.append(data, data[:, 0].reshape(len(data), 1), axis=1)

plt.figure()
ax = plt.subplot(1, 1, 1, polar=True)

ax.plot(angles, data[0], linewidth=1, linestyle='solid', label='Entity 1')
ax.plot(angles, data[1], linewidth=1, linestyle='dashed', label='Entity 2')
ax.plot(angles, data[2], linewidth=1, linestyle='dotted', label='Entity 3')

ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
ax.set_xticks(angles)
ax.set_rlabel_position(0)
ax.set_title('Radar Chart')
ax.tick_params(axis='x', colors='black')
ax.fill(angles, data[0], alpha=0.25)
ax.fill(angles, data[1], alpha=0.25)
ax.fill(angles, data[2], alpha=0.25)
plt.legend()
plt.show()
```

22. Create a Python program to visualize a choropleth map using Plotly, representing the population density of different regions.




```python
import plotly.express as px
import pandas as pd

data = {
    'Region': ['Region A', 'Region B', 'Region C', 'Region D', 'Region E'],
    'Population Density': [150, 300, 120, 450, 200],
    'Code': ['A', 'B', 'C', 'D', 'E']  # Unique codes for each region
}

df = pd.DataFrame(data)

geojson_url = 'https://raw.githubusercontent.com/johan/world.geo.json/master/countries.geo.json'

fig = px.choropleth(df,
                   geojson=geojson_url,
                   featureidkey='properties.name',  # Removed non-breaking space
                   locations='Region',  # Regions in your data
                   color='Population Density',  # Color scale based on population density
                   hover_name='Region',  # Info displayed when hovering over
                   hover_data=['Population Density'],  # Additional hover data
                   color_continuous_scale="Viridis",  # Color scale
                   title="Population Density by Region")

fig.update_geos(fitbounds="locations", visible=False)
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})

fig.show()
```

23. Generate a Python program to create an interactive 3D surface plot using Plotly based on mathematical functions.

```python
import plotly.graph_objects as go
import numpy as np

def f(x, y):
    return x**2 + y**2

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

Z = f(X, Y)

fig = go.Figure(data=[go.Surface(x=X, y=Y, z=Z, colorscale='Viridis')])

fig.update_layout(title='3D Surface Plot',
                  xaxis_title='X',
                  yaxis_title='Y',
                  scene = dict(zaxis_title='Z'))

fig.show()
```

24. Write a Python program to plot a Sankey diagram using Plotly to represent the flow of data between different categories.

```python
import plotly.graph_objects as go

nodes = ["Category A", "Category B", "Category C", "Category D", "Category E"]

sources = [0, 0, 1, 1, 2, 3]
targets = [1, 2, 3, 4, 4, 4]

values = [10, 5, 15, 5, 10, 5]

fig = go.Figure(go.Sankey(
    node=dict(
        pad=15,   # Padding between nodes
        thickness=20,
        line=dict(color="black", width=0.5),  # Border around nodes
        label=nodes
    ),
    link=dict(
        source=sources,  # Sources of the links
        target=targets,  # Targets of the links
        value=values     # Flow values between the nodes
    )
))

fig.update_layout(title_text="Sankey Diagram Example", font_size=10)

fig.show()
```

26. Write a Python program to create an interactive time series plot using Plotly, allowing users to select different date ranges.

27. Generate a Python program to create a treemap using Plotly that displays hierarchical data with different color scales.vvvvvv

```python
import plotly.express as px

data = {
    'Category 1': {'Category 1.1': 10, 'Category 1.2': 15},
    'Category 2': {'Category 2.1': 20, 'Category 2.2': 8}
}

try:
  df = px.sample_codes.treemap(data=data, path_prefix='category_', flat=False)
except AttributeError:

  flat_data = []
  for parent, children in data.items():
      for child, value in children.items():
          flat_data.append({'parent': parent, 'child': child, 'value': value})
  df = pd.DataFrame(flat_data)

fig = px.treemap(df,
                 path=['category_', 'category__'] if hasattr(px, 'sample_codes') else ['parent', 'child'],
                 values='value',
                 color='category_' if hasattr(px, 'sample_codes') else 'parent',
                 color_discrete_map={'category_1': 'green', 'category_2': 'red'},
                 title='Hierarchical Treemap')

fig.show()
```

28. Create a Python program to plot an animated bar chart using Plotly that shows the progression of data over time.

```python
import plotly.express as px
import pandas as pd
import numpy as np

data = {
    'Date': pd.date_range(start='2023-01-01', end='2023-12-31', freq='D'),
    'Category 1': np.random.randint(0, 100, size=365),
    'Category 2': np.random.randint(0, 100, size=365),
    'Category 3': np.random.randint(0, 100, size=365)
}

df = pd.DataFrame(data)

df_melted = df.melt(id_vars='Date', var_name='Category', value_name='Value')

fig = px.bar(df_melted,
             x='Category',
             y='Value',
             color='Category',
             animation_frame='Date',
             animation_group='Category',
             title='Animated Bar Chart')

fig.show()
```

29. Write a Python program to visualize a funnel chart using Plotly, representing the stages of a sales process.

```python
import plotly.graph_objects as go
import pandas as pd

stages = ['Lead', 'Opportunity', 'Proposal', 'Negotiation', 'Closed Won']
values = [1000, 800, 500, 300, 200]

df = pd.DataFrame({'Stage': stages, 'Value': values})

fig = go.Figure(data=[go.Funnel(
    x=df['Value'],
    y=df['Stage'],
    text=df['Stage'],
    name='Sales Funnel',
    marker=dict(
        color='rgb(192, 192, 192)',
        line=dict(color='black', width=1)
    )
)])

fig.update_layout(
    title='Sales Funnel'
)

fig.show()
```

30. Create a Python program to plot a multi-axis line chart using Plotly with different scales on each axis.

```python
import numpy as np
import plotly.graph_objects as go
import pandas as pd

data = {
    'Date': pd.date_range(start='2023-01-01', end='2023-12-31', freq='D'),
    'Stock Price': np.random.randint(100, 200, size=365),
    'Volume': np.random.randint(10000, 50000, size=365)
}

df = pd.DataFrame(data)

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=df['Date'],
    y=df['Stock Price'],
    name='Stock Price',
    yaxis='y1'

fig.add_trace(go.Scatter(
    x=df['Date'],
    y=df['Volume'],
    name='Volume',
    yaxis='y2'
))

fig.update_layout(
    title='Stock Price and Volume Over Time',
    xaxis_title='Date',
    yaxis1=dict(title='Stock Price'),
    yaxis2=dict(title='Volume', overlaying='y1', side='right')
)

fig.show()
```
