**14. What is the difference between a boxplot and a violin plot? How can each be created in Seaborn?**

A **boxplot** summarizes data distribution using quartiles, medians, and outliers, focusing on central tendency and variability. In contrast, a **violin plot** combines a boxplot with a kernel density plot, providing a richer view of the data’s distribution shape and density. Boxplots are ideal for comparing summary statistics, while violin plots offer insights into the underlying data distribution.

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("tips")
data.head()
```

```python
sns.boxplot(x="day",y="total_bill",data=data)

plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.title("Box Plot")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=data)

plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.title("Box Plot")
plt.show()
```

**18. What is a pie chart, and how can it be created using Matplotlib? Provide an example.**

A pie chart is a circular statistical graphic divided into slices to illustrate numerical proportions. Each slice represents a category's contribution to the whole.

Create a pie chart using Matplotlib's plt.pie() function.

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 15]
colors = ['red','blue','yellow','green']
explode = (0.1, 0, 0, 0)

plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', startangle=140)

plt.title('Pie Chart Example')

plt.show()
```

**20. Explain the concept of FacetGrid in Seaborn and its applications.**

`FacetGrid` in Seaborn is a tool for creating a grid of subplots that display different subsets of your data based on one or more categorical variables. Each subplot shows a different segment of the data, allowing for easy comparison across categories.

**Key Points:**
- **Grid Layout**: Arranges multiple subplots in a grid, where each subplot corresponds to a specific category or combination of categories.
- **Customization**: You can apply different plots to each grid cell and customize titles, axis labels, and more.
- **Applications**: Useful for comparing distributions, exploring relationships, and visualizing multi-dimensional data across different categories.

**Example Use Case**: Plotting scatter plots of `total_bill` vs. `tip`, segmented by `time` (Lunch/Dinner) and `sex` (Male/Female) to observe how these factors influence tipping behavior.

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('tips')

g = sns.FacetGrid(data, col='time', row='sex')
g.map(sns.scatterplot, 'total_bill', 'tip')

plt.show()
```

**27. How can you create a subplot with different types of plots in Plotly?**

*   Create Subplots: Use make_subplots to define the grid and subplot types.
*   Add Traces: Use add_trace to add different types of plots to specific subplots.
*   Update Layout: Customize the layout as needed.

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=("Scatter", "Bar", "Histogram", "Box"),
    specs=[[{"type": "scatter"}, {"type": "bar"}],
           [{"type": "histogram"}, {"type": "box"}]]
)

fig.add_trace(go.Scatter(x=[1, 2, 3], y=[4, 5, 6]), row=1, col=1)
fig.add_trace(go.Bar(x=['A', 'B', 'C'], y=[10, 20, 15]), row=1, col=2)
fig.add_trace(go.Histogram(x=[1, 2, 2, 3, 3, 3]), row=2, col=1)
fig.add_trace(go.Box(y=[1, 2, 3, 4, 5]), row=2, col=2)

fig.show()
```

**29. Explain how to add annotations and shapes to a Plotly plot.**

Adding annotations and shapes to a Plotly plot enhances data visualization by providing additional context and highlights.

### **1. Adding Annotations**

Annotations add text labels to specific points or areas in the plot.

**Annotations**: Use `add_annotation` to add text labels with optional arrows.

```python

import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(x=[1, 2, 3], y=[4, 5, 6], mode='markers'))

fig.add_annotation(
    x=2, y=5,         # Coordinates of the annotation
    text="Important Point",  # Annotation text
    showarrow=True,   # Display an arrow pointing to the annotation
    arrowhead=2,      # Arrow style
    ax=0, ay=-40      # Offset of the annotation text
)

fig.show()
```


### **2. Adding Shapes**

Shapes allow you to draw lines, rectangles, circles, etc., on the plot for highlighting areas or adding visual guides.

**Shapes**: Use `add_shape` to draw lines, rectangles, or other shapes on the plot.

```python

fig = go.Figure()

fig.add_trace(go.Scatter(x=[1, 2, 3], y=[4, 5, 6], mode='markers'))

fig.add_shape(
    type="line",               # Shape type
    x0=1, y0=0,                # Start coordinates
    x1=1, y1=6,                # End coordinates
    line=dict(color="RoyalBlue", width=2)  # Line style
)

fig.add_shape(
    type="rect",
    x0=1, y0=4,                # Bottom-left corner
    x1=3, y1=6,                # Top-right corner
    line=dict(color="Green", width=2),  # Rectangle border style
    fillcolor="LightSkyBlue"   # Fill color
)

fig.show()
```

**30. How can you integrate Plotly visualizations into a web application? Provide an example.**

To integrate Plotly visualizations into a web application, you can use libraries like Dash, which is specifically designed for building interactive web applications with Plotly. Here's a brief example using Dash:

```python
pip install dash # Install Dash
```

```python
# Create a Dash Application

import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.graph_objects as go

# Initialize the Dash app
app = dash.Dash(__name__)

# Define a Plotly figure
fig = go.Figure(data=go.Scatter(x=[1, 2, 3], y=[4, 5, 6], mode='markers'))

# Define the layout of the app
app.layout = html.Div([
    html.H1("Plotly with Dash"),
    dcc.Graph(figure=fig)  # Add Plotly graph to the layout
])

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```

1. **Install Dash**: Use `pip` to install the Dash library.
2. **Initialize App**: Create a Dash app instance.
3. **Define Figure**: Create a Plotly figure.
4. **Set Layout**: Define the layout of the web app, including Plotly visualizations using `dcc.Graph`.
5. **Run Server**: Start the Dash server to serve the application.

Dash allows you to integrate Plotly plots into web applications by defining the layout with HTML components and adding Plotly graphs. Running the Dash server makes the web app accessible through a web browser.
